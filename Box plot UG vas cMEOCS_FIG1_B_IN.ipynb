```r
library(ggplot2)
library(ggsignif)
library(ggpubr)
```

```r
data<-read.delim("D:\\Workstation\\cMECOS-upload-data\\UG_sp_num_vs_cMECOS.csv", sep = ',', stringsAsFactors = FALSE, check.names = FALSE)
```

```r
data$variable <- factor(data$variable,levels=c("UNC_NUM_SP","ALL_NUM_SP")) 
```

```r
cols <- c( "#9dbdd2","#f1c299")
```

```r
ggplot(data = data,mapping = aes(x=variable,y=value))+

stat_boxplot(geom="errorbar",width=0.25,aes(x=variable,y=value,color=variable),size=1.5)+
geom_boxplot(outlier.size = 0.1,width=0.3,aes(x=variable,y=value,color=variable),size=1.5)+
stat_compare_means(label = "p.format",size=7,colour="black",vjust=1,paired=T,method = "wilcox.test")+

geom_line(aes(group=People),color="#cdcdcd",alpha=0.7,size=1.5)+
geom_point(size=5,alpha=0.6,aes(x=variable,y=value,color=variable))+

scale_color_manual(values = cols)+
scale_x_discrete(labels = c("UG", "cMECOS"))+
labs(x="",y="Species Richness",color="Type")+ 
theme(panel.background = element_rect(color = '#636363',fill = 'transparent')
      ,strip.text = element_text(size = 14,colour = "black", face = "bold")
      ,strip.background = element_rect(color="transparent",
                                        fill="transparent",
                                        size=0.3, 
                                        linetype="solid"),
      ,legend.title = element_text(size = 12,face = "bold",vjust = 1,
                                   hjust = 0)
      ,legend.text = element_text(size = 12, color="black", face= "bold") 
      ,legend.key.size=unit(1,'cm')          
      ,axis.title.x = element_text(size = 13, vjust = 0.5, 
                                    hjust = 0.5)
      ,axis.title.y = element_text(size = 10, vjust = 1, 
                                    hjust = 0.5,face="bold")
      ,legend.position="none"
      ,axis.text.y= element_text(size=8, color="black", face= "bold")
      ,axis.text.x=element_text(size=8, color="black", face= "bold")
      
    
)
ggsave("add_sp_richness_boxplot.pdf",width = 5,height =6,dpi=600)
```
